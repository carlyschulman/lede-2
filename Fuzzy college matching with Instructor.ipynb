```python
import instructor
from pydantic import BaseModel, Field
from openai import OpenAI
from typing import List
from typing_extensions import Literal
from pydantic.functional_validators import field_validator
from typing import Optional

class School(BaseModel):
    original_name: str = Field(description="Provided name of college")
    school_name: Optional[Literal[school_options]] = Field(description="Most likely name of school")

client = instructor.from_openai(OpenAI(api_key = GPT_API_KEY))
```

```python
from tqdm import tqdm

results = []
colleges = players.college.dropna().unique()

for college in tqdm(colleges):
    result = client.chat.completions.create(
        model="gpt-4o",
        response_model=School,
        messages=[{"role": "user", "content": college}],
        max_retries=3
    )
    
    results.append(result.model_dump())
```

```python

def clean_and_analyze_rankings(df):
    # 1. Check the data type of the ranking column
    print("Initial data type of ranking column:", df['ranking'].dtype)
    
    # 2. Display unique values to see what we're dealing with
    print("\nUnique values in ranking column:")
    print(df['ranking'].unique())
    
    # 3. Convert rankings to numeric, forcing errors to NaN
    df['ranking_numeric'] = pd.to_numeric(df['ranking'], errors='coerce')
    
    # 4. Check for and display any non-numeric values that were converted to NaN
    non_numeric = df[df['ranking_numeric'].isna() & df['ranking'].notna()]
    print(f"\nNon-numeric values converted to NaN: {len(non_numeric)}")
    if len(non_numeric) > 0:
        print(non_numeric[['ranking', 'ranking_numeric']].head())
    
    # 5. Remove any remaining non-numeric values
    df_clean = df.dropna(subset=['ranking_numeric'])
    
    # 6. Convert to integer type
    df_clean['ranking_int'] = df_clean['ranking_numeric'].astype(int)
    
    # 7. Calculate and display statistics
    print("\nRanking statistics:")
    print(f"Mean ranking: {df_clean['ranking_int'].mean():.2f}")
    print(f"Median ranking: {df_clean['ranking_int'].median()}")
    print(f"Min ranking: {df_clean['ranking_int'].min()}")
    print(f"Max ranking: {df_clean['ranking_int'].max()}")
    
    # 8. Display the distribution of rankings
    print("\nRanking distribution:")
    print(df_clean['ranking_int'].value_counts().sort_index().head())
    
    return df_clean
```

```python
# Call the function
cleaned_df = clean_and_analyze_rankings(df)
```

```python
def analyze_team_season(df, team, season):
    # Filter the dataframe for the specific team and season
    team_season_df = cleaned_df[(cleaned_df['team_abbreviation'] == team) & (cleaned_df['season'] == season)]
    
    if team_season_df.empty:
        print(f"No data found for {team} in the {season} season.")
        return
    
    # Calculate required statistics
    mean_rank = team_season_df['ranking_int'].mean()
    median_rank = team_season_df['ranking_int'].median()
    num_players = len(team_season_df)
    
    # Find the player with the highest college rank
    best_college_player = team_season_df.loc[team_season_df['ranking_int'].idxmin()]
    
    # Prepare the results
    results = {
        "Team": team,
        "Season": season,
        "Mean Rank": mean_rank,
        "Median Rank": median_rank,
        "Number of Players with Eligible Colleges": num_players,
        "Best College Player": best_college_player['player_name'],
        "Best Player's College": best_college_player['college'],
        "Best Player's Rank": best_college_player['ranking_int']
    }
    
    # Print the results
    print("\nAnalysis Results:")
    for key, value in results.items():
        print(f"{key}: {value}")
```

```python
def analyze_all_teams_in_season(df, season):
    # Filter the dataframe for the specific season
    season_df = df[df['season'] == season]
    
    if season_df.empty:
        print(f"No data found for the {season} season.")
        return None

    # Get unique teams in this season
    teams = season_df['team_abbreviation'].unique()

    # Initialize a list to store results for each team
    all_team_results = []

    for team in teams:
        team_df = season_df[season_df['team_abbreviation'] == team]
        
        # Calculate required statistics
        mean_rank = team_df['ranking_int'].mean()
        median_rank = team_df['ranking_int'].median()
        num_players = len(team_df)
        
        # Find the player with the highest college rank (lowest number)
        best_college_player = team_df.loc[team_df['ranking_int'].idxmin()]
        
        # Prepare the results
        team_results = {
            "Team": team,
            "Mean Rank": mean_rank,
            "Median Rank": median_rank,
            "Number of Players": num_players,
            "Best College Player": best_college_player['player_name'],
            "Best Players College": best_college_player['college'],
            "Best Players Rank": best_college_player['ranking_int']
        }
        
        all_team_results.append(team_results)

    # Convert results to a DataFrame for easier analysis
    results_df = pd.DataFrame(all_team_results)

    # Sort teams by mean rank
    results_df = results_df.sort_values('Mean Rank')

    # Print summary
    print(f"\nAnalysis Results for {season} Season:")
    print(f"Total teams analyzed: {len(results_df)}")
    print("\nTop 5 teams by mean college rank:")
    print(results_df[['Team', 'Mean Rank', 'Median Rank', 'Number of Players']].head())
    
    print("\nBottom 5 teams by mean college rank:")
    print(results_df[['Team', 'Mean Rank', 'Median Rank', 'Number of Players']].tail())

    # Additional analysis
    print(f"\nTeam with the player from the highest ranked college:")
    best_team = results_df.loc[results_df['Best Players Rank'].idxmin()]
    print(f"Team: {best_team['Team']}")
    print(f"Player: {best_team['Best College Player']}")
    print(f"College: {best_team['Best Players College']}")
    print(f"Rank: {best_team['Best Players Rank']}")

    return results_df
```

```python

# Example usage:
season_to_analyze = "2022-23"  # Replace with the season you want to analyze
results_df = analyze_all_teams_in_season(cleaned_df, season_to_analyze)
results = analyze_all_teams_in_season(cleaned_df, season_to_analyze)
```

```python


all_seasons_teams = []
def analyze_all_seasons_and_teams_df(df):
    # Group by season and team
    grouped = df.groupby(['season', 'team_abbreviation'])
    
    # Iterate through each group
    for (season, team_abbreviation), group in grouped:
        
        # Calculate required statistics
        mean_rank = group['ranking_int'].mean()
        median_rank = group['ranking_int'].median()
        num_players = len(group)
        
        # Find the player with the highest college rank (lowest number)
        best_college_player = group.loc[group['ranking_int'].idxmin()]

        team_results = {
            "Season": season,
            "Team": team_abbreviation,
            "Mean Rank": mean_rank,
            "Median Rank": median_rank,
            "Number of Players": num_players,
            "Best College Player": best_college_player['player_name'],
            "Best Players College": best_college_player['college'],
            "Best Players Rank": best_college_player['ranking_int']
        }

        all_seasons_teams.append(team_results)

    # Convert results to a DataFrame for easier analysis
    all_seasons_teams_df = pd.DataFrame(all_seasons_teams)

    # Sort teams by mean rank
    all_seasons_teams_df = all_seasons_teams_df.sort_values('Median Rank')

    return all_seasons_teams_df

#print(f"Median ranking: {df_clean['ranking_int'].median()}")
```
